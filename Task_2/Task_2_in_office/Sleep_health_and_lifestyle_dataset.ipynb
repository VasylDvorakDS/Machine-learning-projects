#Корреляционный анализ:

- это статистический метод, который используется для изучения взаимосвязи между двумя или более переменными. Главная цель корреляционного анализа - определить, насколько сильно и в каком направлении связаны эти переменные.

Основные понятия корреляционного анализа:
1. *Коэффициент корреляции* - это числовое значение, которое показывает степень взаимосвязи между двумя переменными. Коэффициент корреляции может колебаться от -1 до 1. Значения близкие к 1 указывают на сильную положительную корреляцию, значения близкие к -1 указывают на сильную отрицательную корреляцию, а значение равное 0 означает отсутствие корреляций.

2. *Положительная корреляция*. Если значения одной переменной возрастают, то значения другой переменной также увеличиваются.

3. *Отрицательная корреляция*. Если значения одной переменной увеличиваются, то значения другой переменной уменьшаются.

4. *Нулевая корреляция*: отсутствие взаимосвязи между переменными.

Корреляционный анализ позволяет определить, есть ли статистически значимая связь между переменными, и какой ее характер. Важно помнить, что корреляция не всегда означает причинно-следственную связь между переменными - она лишь отражает степень и направление их взаимосвязи.

Корреляционный анализ широко используется в многих областях, включая науку, экономику, медицину, социологию и др., для выявления закономерностей и анализа данных.

#**Коэффициенты корреляции**

Существует несколько коэффициентов корреляции, которые используются для измерения степени взаимосвязи между переменными. Основные типы коэффициентов корреляции включают:

1. *Коэффициент Пирсона (Pearson's r)* - Самый распространенный и широко используемый коэффициент корреляции.
- Измеряет линейную зависимость между двумя количественными переменными.
- Может колебаться от -1 (полная отрицательная корреляция) до 1 (полная положительная корреляция).

2. *Коэффициент Спирмена (Spearman's rho)*. - Используется для измерения силы монотонной взаимосвязи между двумя переменными.
- Не требует нормального распределения данных и может использоваться с ранговыми данными.

3. *Коэффициент Кендалла (Kendall's tau)*. - Также измеряет силу монотонной взаимосвязи между переменными.
- Чувствителен к выбросам в данных и показывает, насколько часто конкретные значения одной переменной связаны с конкретными значениями другой переменной.

Коэффициент корреляции Пирсона выбирается для анализа связи между физической активностью и качеством сна по нескольким причинам:

1. Пирсона требует непрерывных переменных: И физическая активность (если мы рассматриваем её как количество минут или часов упражнений в неделю), и самооценка качества сна (которую можно выразить через числовую шкалу, например, от 1 до 10) являются количественными непрерывными переменными. Это удовлетворяет ключевое требование для использования корреляции Пирсона.

2. Оценка линейной взаимосвязи: Коэффициент корреляции Пирсона измеряет степень и направление линейной связи между двумя переменными. Если предполагается, что между уровнем физической активности и качеством сна существует прямая (чем больше активность, тем лучше качество сна) или обратная (чем больше активность, тем хуже качество сна) связь, корреляция Пирсона является подходящим статистическим инструментом для оценки этой связи.

3. Количество данных: Имеется достаточное количество данных (374 непрерывных наблюдений), что дает статистическую мощность для выявления значимой корреляции.

###Вычисляем коэффициент корреляции Пирсона

```python
import scipy.stats as stats
# Вычисляем коэффициент корреляции Пирсона
correlation_coefficient, p_value = stats.pearsonr(df['Physical Activity Level'], df['Quality of Sleep'])

print(f"Коэффициент корреляции Пирсона: {correlation_coefficient}")
print(f"P-значение: {p_value}")
```

```python
df['Physical Activity Level'].corr(df['Quality of Sleep'], method = 'pearson')
```

###Визуализация корреляции

```python
import matplotlib.pyplot as plt

# Построение графика рассеивания
plt.figure(figsize=(10, 6))
plt.scatter(df['Physical Activity Level'], df['Quality of Sleep'], alpha=0.5)
plt.title('Связь между физической активностью и качеством сна')
plt.xlabel('Уровень физической активности')
plt.ylabel('Качество сна')
plt.grid(True)
plt.show()
```

Коэффициент корреляции Пирсона равен 0.19289645493975327, и это означает, что между уровнем физической активности и качеством сна есть положительная связь, но она довольно слабая, поскольку коэффициент близок к 0.

P-значение равное 0.00017454849254783864 говорит о статистической значимости результатов. Поскольку это значение меньше стандартного порога в 0.05 для статистической значимости, мы можем сказать, что существует статистически значимая корреляция между уровнем физической активности и качеством сна, пусть даже и небольшая.

Исходя из полученных данных, мы можем сделать вывод, что хотя уровень физической активности и качество сна как-то связаны, эта связь не является сильной. Необходимо также учитывать, что корреляция не подразумевает причинно-следственной связи, а лишь указывает на наличие статистической взаимосвязи. Для определения причин и эффектов необходимо проведение более детального анализа, возможно с использованием экспериментальных или квазиэкспериментальных методов.

###Вычисление коэффициента Спирмена

```python
from scipy.stats import spearmanr

# Пример для Stress Level и Quality of Sleep
coef, p = spearmanr(df['Stress Level'], df['Quality of Sleep'])
print(f"Коэффициент корреляции Спирмена между уровнем стресса и качеством сна: {coef}, P-value: {p}")

# Пример для Physical Activity Level и Quality of Sleep
coef, p = spearmanr(df['Physical Activity Level'], df['Quality of Sleep'])
print(f"Коэффициент корреляции Спирмена между уровнем физической активности и качеством сна: {coef}, P-value: {p}")

```

###Визуализация

   - Цель анализа: изучить взаимосвязь между возрастом участников и их самооценкой качества сна.
   Гипотеза: существует монотонная зависимость между возрастом и качеством сна, так что с возрастом качество сна изменяется определенным образом (улучшается или ухудшается).
   - Цель исследования: Оценить степень монотонной взаимосвязи между уровнем стресса и частотой сердцебиения у участников исследования.


###Коэффициент корреляции Кэнделла между возрастом (`"Age"`) и качеством сна (`"Quality of Sleep"`).


Обоснование:  
Коэффициент корреляции Кэнделла выбран для анализа, так как он позволяет изучить монотонные взаимосвязи между количественной (возраст) и порядковой (качество сна) переменными без предположения о линейности связи или равномерности шкал. Визуализация данных с помощью графика рассеивания поможет наглядно определить, как распределяются данные и подтверждает ли визуальное наблюдение результаты расчета коэффициента Кэнделла.

```python
from scipy.stats import kendalltau


# Расчет коэффициента Кэнделла
tau, p_value = kendalltau(df["Age"], df["Quality of Sleep"])
print(f"Коэффициент корреляции Кэнделла: {tau}, P-value: {p_value}")

# Визуализация
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Age", y="Quality of Sleep", data=df)
plt.title('Взаимосвязь возраста и качества сна')
plt.xlabel('Возраст')
plt.ylabel('Качество сна')
plt.show()
```

Исходя из полученного значения коэффициента корреляции Кэнделла (0.30899361248680174) и крайне низкого p-value (3.592373043767236e-15), мы можем сделать вывод о наличии статистически значимой положительной монотонной взаимосвязи между возрастом участников и качеством их сна. Это означает, что с увеличением возраста качество сна участников исследования имеет тенденцию к улучшению или оценивается участниками как более высокое.

Следует отметить, что коэффициент корреляции Кенделла выше 0, но не достаточно высок, чтобы говорить о сильной взаимосвязи, хотя взаимосвязь и значима. Это указывает на то, что возраст может влиять на качество сна, но этот фактор не является единственным или определяющим качество сна, и есть и другие факторы, которые также следует учитывать.

### Анализ связи между уровнем стресса (`"Stress Level"`) и частотой сердцебиения (`"Heart Rate"`) с использованием коэффициента корреляции Кэнделла

###Обоснование выбора коэффициента Кэнделла:

Коэффициент Кэнделла выбран для анализа взаимосвязи между уровнем стресса и частотой сердцебиения, так как он эффективен для определения монотонных зависимостей между порядковыми или количественными переменными, особенно в случаях, когда предполагается нелинейная связь или присутствуют привязки в данных. Этот метод позволяет учесть как величину изменений, так и направление взаимосвязи между переменными.

```python
# Расчет коэффициента корреляции Кэнделла
tau, p_value = kendalltau(df["Stress Level"], df["Heart Rate"])
print(f"Коэффициент корреляции Кэнделла: {tau}, P-value: {p_value}")

# Визуализация взаимосвязи
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Stress Level", y="Heart Rate", data=df)
plt.title('Связь между уровнем стресса и частотой сердцебиения')
plt.xlabel('Уровень стресса')
plt.ylabel('Частота сердцебиения')
plt.show()
```

Результаты анализа показывают коэффициент корреляции Кэнделла равный 0.7138781434027444 с p-value 1.3762691073665193e-69. Это указывает на сильную положительную монотонную связь между уровнем стресса и частотой сердцебиения. Значение коэффициента корреляции близко к 1, что говорит о том, что с увеличением уровня стресса, как правило, увеличивается и частота сердцебиения, что логично с физиологической точки зрения, поскольку стрессовые состояния зачастую приводят к учащению сердцебиения.

Крайне низкое значение p-value (значительно меньше стандартного порога 0.05) свидетельствует о высокой статистической значимости полученных результатов. Это означает, что вероятность того, что наблюдаемая взаимосвязь между уровнем стресса и частотой сердцебиения является случайной, чрезвычайно мала.

### Выводы
Опираясь на эти результаты, можно сделать вывод, что уровень стресса имеет значительное и прямое влияние на частоту сердцебиения у исследуемой группы людей. Это подтверждает гипотезу о тесной связи физиологических показателей человека со степенью его психоэмоциональной напряженности. Важно отметить, что такой анализ может быть полезен в медицинских и психологических исследованиях для разработки стратегий по снижению стресса и контролю сердечно-сосудистой системы.
